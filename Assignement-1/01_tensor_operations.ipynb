## Function 1 - `torch.normal`

n probability theory, a normal (or Gaussian or Gauss or LaplaceGauss) distribution is a type of continuous probability distribution for a real-valued random variable. The general form of its probability density function is:
## $$f(x) = \frac{1}{\sigma\sqrt{2\pi}}e^{\frac{-1}{2}(\frac{x-\mu}{\sigma})^2}$$

This pdf is integrated via `torch.normal(mean, std, *, generator=None, out=None)`.
It returns a tensor of random numbers drawn from separate normal distributions whose mean and standard deviation are given.

The mean is a tensor with the mean of each output elements normal distribution

The std is a tensor with the standard deviation of each output elements normal distribution

The shapes of mean and std dont need to match, but the total number of elements in each tensor need to be the same.



```python
t1 = torch.normal(mean=torch.arange(1.,41), std=torch.arange(-2, 2, 0.1))
t1
```

The result is a tensor of random numbers drawn from separate normal distributions.

```python
t2 = torch.normal(mean=0., std=torch.arange(-2, 2, 0.1))
t2
```

The result is a tensor of random of numbers drawn from separate normal distributions, but with the $\mu = 0$.  
Let's try if this function can integrate a central limit theorem when the standard normal distribution converges to N(0,1).
with $\mu = 0$ and $\sigma = 1$

```python
t3 = torch.normal(mean=0,std=1)
```

As it was expected it doesn't work.For this to work, you have to invoke the tensor.normal_() method.

When a random variable X follows a normal distribution, it is said to be Gaussian or normal and it is usual to use the notation with the variance $\sigma$:
## $$X \sim N(\mu,\sigma^2)$$  
Among the probability laws, the normal laws take a special place thanks to the central limit theorem.

## Function 2 - `tensor.normal_()`

In-place version of `torch.normal()`.Fills self tensor with elements samples from the normal distribution parameterized by  and .

```python
t4 = torch.empty(40)
t5 = t4.normal_(mean=0,std=1)
t5
```

We can see that this function allows us to easily converge towards the central limit theorem, unlike the previous function.

```python
t4.normal_(0.05, 1)
```

## Function 3 - `torch.poisson(input *, generator=None)`

The Poisson distribution is popular for modeling the number of times an event occurs in an interval of time or space.

A discrete random variable X is said to have a Poisson distribution with parameter  > 0 if for k = 0, 1, 2, ..., the probability mass function of X is given by:
## $$f(k,\lambda) = P(X=k) = \lambda^k\frac{e^{-\lambda}}{k!}$$  

`torch.poisson` returns a tensor of the same size as input with each element sampled from a Poisson distribution with rate parameter given by the corresponding element in input i.e.,
## $$out_i \sim Poisson(input_i)$$

```python
rates = torch.rand(3, 3) * 6  # rate parameter between 0 and 6 with 3 row and 3 col 
rates
```

```python
torch.poisson(rates)
```

A input tensor contain rate parameter between 0 and 6 with 3 rows and 3 columns, returns a output from a poission distribution.

```python
rates = torch.rand(5, 4) * 7  # rate parameter between 0 and 7 with 5 row and 4 col 
torch.poisson(rates)
```

A input tensor contain rate parameter between 0 and 7 with 5 rows and 4 columns, returns a output from a poission distribution

```python
rates = torch.rand(5, 6) * 9  # rate parameter between 0 and 9 with 5 row and 6 col 
torch.poisson(rates)
```

## Function 4 - `torch.bernoulli(input, *, generator=None, out=None)`

the Bernoulli distribution or Bernoulli's law, is a discrete probability distribution, which takes the value 1 with the probability p and 0 with the probability q = 1 - p. In other words,
## $$P(X=x) = p^x(1-p)^{1-x}, x \in \{0,1\}$$

`torch.bernoulli` function draws binary random numbers (0 or 1) from a Bernoulli distribution.

```python
t = torch.tensor([[0.1,0.55,0.79],[0.09,0.48,0.33],[0.30,0.39,0.9]])
```

create a tensor

```python
#Draws a binary random nuber(0 or 1)
torch.bernoulli(t)
```

We create a custom tensor and passed to bernoulli function it return a binary number (0 or 1).

```python
torch.bernoulli(torch.ones(5,5))
```

We can see that the probability of drawing "1" is 1.

## Function 5 - `exponential_(lambd=1, *, generator=None)`

The exponential distribution is the probability distribution of the time between events in a Poisson point process, i.e., a process in which events occur continuously and independently at a constant average rate.  
`exponential_`function returns a tensor filled with random values from the exponential distribution parameterized by `lambd`.

```python
t = torch.empty(5,3)
t
```

Create an initialized a tensor of size 5 rows 3 cols

```python
t.exponential_()
```

This method modifies the `tensor` t in place by filling it with values from this exponential distribution and then returns it.

```python
t.exponential_(0.5)
```

Replaces the old values in the matrix by random values from an exponential distribution that has a rate-parameter of 0.5, i.e. average time/space between events (successes) that follow a Poisson Distribution is 0.5.
